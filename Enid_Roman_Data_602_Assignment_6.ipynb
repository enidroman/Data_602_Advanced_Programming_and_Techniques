1. Import a “dirty” dataset from the internet into your environment. (Missing values, improper coding of columns, etc.)
2. Clean or “tidy” the data when loading into python using pandas.

```python
dirtydf = pd.read_csv('https://raw.githubusercontent.com/data602sps/datasetspractice/main/atu_dirty.csv')
```

```python
# Preview dataset, dirtydf.
dirtydf
```

```python
# Changed columns names as to https://curriculum.idsucla.org/unit1/lab1f/, where I got information on the data. 
cleanerdf = dirtydf.rename({'caseid':'Case Id', 'V1':'Age', 'V2':'Gender', 'V3':'Employ status', 'V4': 'Disability',	'V5':'Amt of Sleep',
                            'V6':'Time in HW',	'V7':'Amt Socializing'}, axis='columns') 
cleanerdf                    
```

```python
# Dropped Unamed: 0 column. 
newdf = cleanerdf.drop(columns=['Unnamed: 0'])
newdf
```

```python
#Checking # of columns
len(newdf.columns)
```

```python
# Checking # of rows. 
newdf.count()[0]
```

```python
# Datatype  info. 
newdf.info()
```

```python
# Checking if there are any missing value on the entire dataset
newdf.isnull().sum()
```

```python
# Replaced No answer with N/A
newdf.replace(to_replace="No answer",
           value="N/A")
```
