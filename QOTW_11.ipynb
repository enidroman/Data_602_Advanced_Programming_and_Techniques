1. How do you plot a histogram in Seaborn?  

The histplot function by would default take the numerical data variable as an argument, and the output is the histogram of the passed values.

```python
# libraries & dataset
import seaborn as sns
import matplotlib.pyplot as plt
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")
df = sns.load_dataset("iris")
sns
# Plotting a histogram in seaborn
sns.histplot(data=df, x="sepal_length")
plt.show()
```

2. Plot a histogram with NAs dropped.

dropna() method is used to drop missing values in a Pandas DataFrame. 

```python
# Plotting a histogram with NAs dropped
sns.distplot(df['sepal_length'].dropna())
```

3. How do you set the color for a histogram?

Seaborn's distplot() makes the histogram filling the bars in blue. We can manually change the histogram color using the color argument inside distplot() function. 

```python
# Setting a color for plot
sns.distplot(df['sepal_length'].dropna(), bins=50, kde=False, color="g")
```

4.  What type of plot would allow you to compare two continuous features?  Give an example of code.

Scatter plot are used to display the relationship between two continuous features x and y.

```python
sns.scatterplot(x='sepal_length', y='petal_length', data=df, hue = 'species')
```

5. Give example of a correlation plot.

Correlation heatmaps can be used to find potential relationships between variables and to understand the strength of these relationships. In addition, correlation plots can be used to identify outliers and to detect linear and nonlinear relationships. The color-coding of the cells makes it easy to identify relationships between variables at a glance. Correlation heatmaps can be used to find both linear and nonlinear relationships between variables.

```python
# Plotting a correlation plot
corr = df.corr()
sns.heatmap(corr, 
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values)
corr
```

6. Change the figure size of your plot(s).

The figsize() attribute can be used when you want to change the default size of a specific plot.
